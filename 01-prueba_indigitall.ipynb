# Mediciones de contaminación de Madrid en el año 2018
Leer y tratar los datos

## 1. Lectura
Lectura de los datos alojados en formato csv en el directorio

```python
from pathlib import Path
import pandas as pd


dirname = './anio201810'
 
paths = Path(dirname).glob('**/*.csv',)

dfs_list = [None]*12
for path in paths:
    df = pd.read_csv(path, sep=";")
    dfs_list[int(df["MES"].median())-1] = df

complete_df = pd.concat(dfs_list, axis=0, ignore_index=True)
complete_df
```

## 2. Limpieza de datos
Selección de datos referentes a NO2 y limpieza de columnas

```python
no2_df = complete_df[complete_df["MAGNITUD"]==8]
no2_df
```

```python
no2_df = no2_df.drop(columns=["PROVINCIA", "MUNICIPIO", "PUNTO_MUESTREO"], axis=1)
no2_df
```

## 3. Tratamiento de datos
De cara a poder hacer un análisis mejor de los datos facilitados es mejor tener dispuestos los valores de las columnas H01...H24 en una sola, de la misma manera que las columnas V01...V24.

```python
h_columns = [f"H{i:02d}" for i in range(1,25)]
v_columns = [f"V{i:02d}" for i in range(1,25)]

#melt df by HXX values
df_no2_hmelted = no2_df.melt(id_vars=["ESTACION","MES","DIA"], value_vars=h_columns, var_name='HORA', value_name='MEDICION')
df_no2_hmelted
```

```python
#melt df by VXX columns
df_no2_vmelted = no2_df.melt(id_vars=["ESTACION","MES","DIA"], value_vars=v_columns, var_name='V', value_name='VALIDADO')
df_no2_vmelted
```

```python
final_df = pd.concat([df_no2_hmelted, df_no2_vmelted["VALIDADO"]], axis=1)
final_df
```

## 4. GUARDAR DATOS
Guardar los datos en formato csv ya que permite leerlo con diferentes programas, incluso con diferentes lenguajes de programación e igualmente se pueden exportar facilmente a Excel.

```python
final_df.to_csv("madrid_contaminacion_2018.csv", sep=";")
```
